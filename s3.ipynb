```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_performance.csv")
df
```

```python
df.isnull().sum()
```

```python
meanAttendance = df['Attendance Rate'].mean()
print("Mean of Attendance=", meanAttendance)
```

```python
mean=df['Study Hours per Week'].mean()
print("Mean of Study Hours per Week= ", mean)
```

```python
df.boxplot()
plt.title('Box plot')
plt.show()
```

```python
plt.boxplot(df['Attendance Rate'])
plt.show()
```
