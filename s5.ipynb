```python
import pandas as pd
```

```python
data = {
    "Education":["Graduate","Graduate","Postgraduate","Postgraduate","Undergraduate","Undergraduate"],
    "Income":[30000,32000,40000,45000,34000,36000]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
summary_statistics = df.groupby("Education")["Income"].agg(
    Mean = "mean",
    Median = "median",
    Minimum = "min",
    Maximum = "max",
    Standard_Deviation = "std"
)
```

```python
summary_statistics
```

```python
income_list = df.groupby("Education")["Income"].agg(list).to_dict()
display(income_list)
```
